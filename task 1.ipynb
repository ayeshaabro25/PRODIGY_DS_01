```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_622630.csv', skiprows=4)
```

```python
selected_countries = population_data[['Country Name', '2022']].head(20).dropna()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
barplot = sns.barplot(
    x=selected_countries['2022'],
    y=selected_countries['Country Name'],
    palette='coolwarm'
)
plt.xlabel('Population in 2022')
plt.ylabel('Country Name')
plt.title('Population Distribution by Country in 2022 (First 20 Countries)')
plt.tight_layout()
plt.show()
```
